```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import numpy as np
# Nama file CSV yang akan dibaca
csv_file_name = 'coba2.csv'

# Baca data dari file CSV menggunakan pandas
data = pd.read_csv(csv_file_name)

plt.figure(figsize=(12, 6))
plt.plot(data['Time'], data['SpeedMotor'],color= 'blue', label='Kecepatan Motor', linewidth=2)
plt.xlabel('Time (seconds)')
plt.ylabel('Speed')
plt.title('Speed Response motor 2')
plt.grid(False)
plt.legend()

speedMotor = data['SpeedMotor']
rata_rata_speedMotor = speedMotor.mean()
print(f'Rata-rata Kecepatan Motor:{rata_rata_speedMotor:.5f}')

```
